```python
# Import libraries
from __future__ import print_function
import sys
from datetime import datetime
import time
import pandas as pd
```

```python
# define variables
file_in = "Resources/Police_Department_Incident_Reports__2018_to_Present.csv"
file_out_dupes ="Resources/SFPD_duplicates.csv"
# read in CSV
df = pd.read_csv(file_in)
df.head()
```

```python
# Print length of dataframe
print(len(df))
```

```python
# assign field name
#fl_incid = 'Incident Number'
fl_incid = 'Incident ID'
```

```python
key_field = set(df[fl_incid].values)
header = True
```

```python
# concatenate values in instances where field_name value > 1
data = pd.concat(g for _, g in df.groupby(fl_incid) if len(g) > 1)
if len(data) > 1:
    print('Found duplicates')
    data.to_csv(file_out_dupes, mode='w', header=header, index=False)
else:
    pass
```

```python
dupe_df = pd.read_csv(file_out_dupes)
#dupe_df.set_index('IncidntNum', inplace=True)
dupe_df.head()
```
